```python
ode_cols = ['Tranche age']
ohe_cols = ['Sex','Accès ressources financières','Participé à des compétitions','Compte déjà entreprendre','Participé à des compétitions','Avoir un mentor','Travaillé sur des projets','Suivi formations entrepreneuriat','Principale motivation','Filière']
```

```python
X=df.drop(['Target'],axis=1)
y=df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify=y,random_state=21)
```

```python
ordinal_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ord',OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
```

```python
ohe_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('one-hot',OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
```

```python
col_trans=ColumnTransformer(transformers=[
    ('ord_pipeline', ordinal_pipeline, ode_cols),
    ('ohe_pipeline', ohe_pipeline, ohe_cols),
    ('passthrough', 'passthrough', ['Année d\'étude', 'Compétences gestion financière','Compétences leadership','Prise risques financiers', 'Perséverence', 'Motivation à entreprendre'])
    ],
    remainder='drop',
    n_jobs=-1)
```

```python
rfc=RandomForestClassifier()
```

```python
param_grid={
    'n_estimators':[100,150, 200],
    'min_samples_split':[5,10, 15],
    'max_depth':[8,9,10,15, 20],
    'min_samples_leaf':[1,2, 4],
    'criterion':['gini', 'entropy'],
}
```

```python
CV_rfc=GridSearchCV(estimator=rfc, param_grid=param_grid, cv=StratifiedKFold(n_splits=5))
```

```python
pipefinalrfc=make_pipeline(col_trans, CV_rfc)
pipefinalrfc.fit(X_train,y_train)
```

```python
print(CV_rfc.best_params_)
print(CV_rfc.best_score_)
```

```python
knn=KNeighborsClassifier()
```

```python
param_grid={
    'n_neighbors':[3,5,7,9,11],
    'weights':['uniform','distance'],
    'algorithm':['auto','ball_tree', 'kd_tree','brute'],
    'p':[1, 2],
}
```

```python
CV_knn=GridSearchCV(estimator=knn, param_grid=param_grid, cv=StratifiedKFold(n_splits=5))
```

```python
pipefinalknn=make_pipeline(col_trans, CV_knn)
pipefinalknn.fit(X_train,y_train)
```

```python
print(CV_knn.best_params_)
print(CV_knn.best_score_)
```

```python
svc=SVC()
```

```python
param_grid={
    'C':[100,10,1.0,0.1,0.001,0.001],
    'kernel':['linear','poly','rbf','sigmoid'],
}
```

```python
CV_svc=GridSearchCV(estimator=svc, param_grid=param_grid, cv=StratifiedKFold(n_splits=5))
```

```python
pipefinalsvc=make_pipeline(col_trans, CV_svc)
pipefinalsvc.fit(X_train,y_train)
```

```python
print(CV_svc.best_params_)
print(CV_svc.best_score_)
```

```python
lr=LogisticRegression()
```

```python
param_grid={
    'var_smoothing':[0.00000001,0.000000001,0.00000001],
}
```

```python
y_pred_rfc= pipefinalrfc.predict(X_test)
y_pred_knn= pipefinalknn.predict(X_test)
y_pred_svc= pipefinalsvc.predict(X_test)
y_pred_lr= pipefinallr.predict(X_test)
y_pred_gnb= pipefinalgnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

# Fonction pour afficher les métriques d'évaluation
def evaluate_model(y_test, y_pred, model_name):
    print(f"Évaluation du modèle: {model_name}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
    print(f"Classification Report:\n{classification_report(y_test, y_pred)}")
    print("\n")

# Évaluer chaque modèle
evaluate_model(y_test, y_pred_rfc, "Random Forest Classifier")
evaluate_model(y_test, y_pred_knn, "K-Nearest Neighbors")
evaluate_model(y_test, y_pred_svc, "Support Vector Classifier")
evaluate_model(y_test, y_pred_lr, "Logistic Regression")
evaluate_model(y_test, y_pred_gnb, "Gaussian Naive Bayes")
```
